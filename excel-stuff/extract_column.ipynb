```python
import pandas as pd
```

```python
b = pd.read_csv('C:/Users/prayme/Downloads/10.21 교수간담회 로그 기록.CSV', encoding='utf8')
b
```

```python
# extract columns
c = b.filter(items=['Name', 'Duration(Minutes)'])
c
```

```python
DICT = {}
for idx, row in c.iterrows():
    try:
        DICT[row['Name']] += row['Duration(Minutes)']
    except:
        DICT[row['Name']] = row['Duration(Minutes)']
```

```python
DICT2 = {
    'Name': list(DICT.keys()),
    'Duration(Minutes)': list(DICT.values())
}
```

```python
df = pd.DataFrame.from_dict(DICT2)
```

```python
df.to_csv('./save.csv', encoding='utf-8-sig')
```
